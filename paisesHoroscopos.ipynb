# Leo el fichero de paises

```pyspark3
df2 = spark.read.format("csv").load("/data/paises_estandar.csv", sep='\t', header=True, inferSchema=True)
```

# Creo la estructura de la taba de paises

```pyspark3
%%sql -q
CREATE TABLE IF NOT EXISTS paises ( 
                    Continente string, 
                    country string )
```

# Guardo los datos de paises en la tabla

from pyspark.sql import DataFrameWriter

dfw = DataFrameWriter(df2)
dfw.insertInto('paises', overwrite=True)

# Compruebo que se guardaron los datos en la tabla

```pyspark3
paisesdf = spark.sql("""
SELECT * 
FROM paises 
""")
paisesdf.show()
```

# Leo el fichero de horoscopos

```pyspark3
dfHoroscopos = spark.read.load("/data/horoscopos.csv",
                     format="csv", sep=",", inferSchema="true", header="true")
```

# Creo la estructura de la tabla

```pyspark3
%%sql -q
CREATE TABLE IF NOT EXISTS horoscopos ( 
                    nombre string, 
                    fechaDesde string, 
                    fechaHasta string)
```

# Guardo los datos de horoscopos en la tabla

```pyspark3
dfw2 = DataFrameWriter(dfHoroscopos)
dfw2.insertInto('horoscopos', overwrite=True)
```

horoscoposdf = spark.sql("""
