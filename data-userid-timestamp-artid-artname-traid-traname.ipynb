```pyspark3
from pyspark.sql import *
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark import SparkContext
sc = SparkContext.getOrCreate()
sqlContext = SQLContext(sc)




import pandas as pd
```

# Creamos el esquema del dataframe que vamos a crear ya que no tiene encabezamiento

```pyspark3

schemaCabeceras = StructType([\
    StructField("userid", StringType(), True), \
    StructField("timestamp", DateType(), True), \
    StructField("musicbrainz_artist_id", StringType(), True), \
    StructField("artist_name", StringType(), True), \
    StructField("musicbrainz_track_id", StringType(), True), \
    StructField("track_name", StringType(), True) \
])
```

# Carga del archivo con la agregación de las cabeceras

```pyspark3
df = spark.read.format("csv").load("/data/userid-timestamp-artid-artname-traid-traname.tsv", sep='\t',names=["userid","timestamp","musicbrainz_artist_id","artist_name","musicbrainz_track_id","track_name"], schema = schemaCabeceras )
```

```pyspark3
#Comprobamos que ha cogido el esquema de forma correcta
df.printSchema()
```

# Comprobamos si hay nulos en las columnas userid y timestamp

```pyspark3
df.filter(df.userid.isNull()).count()
```

```pyspark3
df.filter(df.timestamp.isNull()).count()
```

# Guardar datos en CSV

```pyspark3
df.write.csv('wasb:///data/datos_musica/Datos_musica.csv', sep=";", header = True)
```
